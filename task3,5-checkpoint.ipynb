```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('test1.mp4')

x1,y1 = 0,0
x2,y2 = 0,0
a,b = 0,0
c,d = 0,0
```

```python
while cap.isOpened():
    n=0
    ret , frame =cap.read()
    x,y = frame.shape[0:2]
    black_image = np.zeros((x,y))
    gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    ret ,threshold =cv2.threshold(gray , 121 ,255 , cv2.THRESH_BINARY)
    median_blur = cv2.medianBlur(gray ,5)
    
   
    def cent(contours , image):
        global n,y1,y2,a,b,c,d
        if cv2.contourArea(contours)>1000:
            cent_moment = cv2.moments(contours)
            centroid_x = int(cent_moment['m10']/cent_moment['m00'])
            centroid_y = int(cent_moment['m01']/cent_moment['m00'])
            cv2.circle(image,(centroid_x,centroid_y),4,(255,0,0),-1)
            cv2.putText(image ,f"({str(centroid_x)},{str(centroid_y)})", (centroid_x,centroid_y), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
            if centroid_x > 420:
                x1=centroid_x
                y1=centroid_y
                speed1y =int((y1-a)/0.02)
                speed1x =int((x1-c)/0.02)
                if speed1y > 0 :
                    cv2.line(image , (centroid_x,y1),(centroid_x,y1+100),(255,0,0),1)
                    cv2.line(image , (centroid_x,y1+100),(centroid_x+20,y1+80),(255,0,0),1)
                    cv2.line(image , (centroid_x,y1+100),(centroid_x-20,y1+80),(255,0,0),1)
                    cv2.putText(image ,'speedy('+str(speed1y)+')', (centroid_x,centroid_y+100), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                elif speed1y<0 :
                    cv2.line(image , (centroid_x,y1),(centroid_x,y1-100),(255,0,0),1)
                    cv2.line(image , (centroid_x,y1-100),(centroid_x+20,y1-80),(255,0,0),1)
                    cv2.line(image , (centroid_x,y1-100),(centroid_x-20,y1-80),(255,0,0),1)
                    cv2.putText(image ,'speedy('+str(speed1y)+')', (centroid_x,centroid_y-100), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
#                 elif speed1y == 0:
#                     pass
                
                
                if speed1x >0:
                    cv2.line(image , (centroid_x,centroid_y), (centroid_x+100,centroid_y),(255,0,0),1)
                    cv2.line(image , (centroid_x+100,centroid_y), (centroid_x+80,centroid_y+20),(255,0,0),1)
                    cv2.line(image , (centroid_x+100,centroid_y), (centroid_x+80,centroid_y-20),(255,0,0),1)
                    cv2.putText(image ,'speedx('+str(speed1x)+')', (centroid_x+100,centroid_y), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                elif speed1x <0:
                    cv2.line(image ,(centroid_x,centroid_y),(centroid_x-100 ,centroid_y),(255,0,0),1)
                    cv2.line(image ,(centroid_x-100,centroid_y),(centroid_x-80 ,centroid_y+20),(255,0,0),1)
                    cv2.line(image ,(centroid_x-100,centroid_y),(centroid_x-80 ,centroid_y-20),(255,0,0),1)
                    cv2.putText(image ,'speedx('+str(speed1x)+')', (centroid_x-100,centroid_y), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                    
                c=x1
                a=y1
                
                
            elif centroid_x < 420:
                x2=centroid_x
                y2=centroid_y
                speed2y =int((y2-b)/0.02)
                speed2x =int((x2-d)/0.02)
                
                if speed2y >0 :
                    cv2.line(image , (centroid_x,y2),(centroid_x,y2+100),(255,0,0),1)
                    cv2.line(image , (centroid_x,y2+100),(centroid_x+20,y2+80),(255,0,0),1)
                    cv2.line(image , (centroid_x,y2+100),(centroid_x-20,y2+80),(255,0,0),1)
                    cv2.putText(image ,'speedy('+str(speed2y)+')', (centroid_x,centroid_y+100), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                elif speed2y <0 :
                    cv2.line(image , (centroid_x,y2),(centroid_x,y2-100),(255,0,0),1)
                    cv2.line(image , (centroid_x,y2-100),(centroid_x+20,y2-80),(255,0,0),1)
                    cv2.line(image , (centroid_x,y2-100),(centroid_x-20,y2-80),(255,0,0),1)
                    cv2.putText(image ,'speedy('+str(speed2y)+')', (centroid_x,centroid_y-100), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                    
                    
                if speed2x >0:
                    cv2.line(image , (centroid_x,centroid_y), (centroid_x+100,centroid_y),(255,0,0),1)
                    cv2.line(image , (centroid_x+100,centroid_y), (centroid_x+80,centroid_y+20),(255,0,0),1)
                    cv2.line(image , (centroid_x+100,centroid_y), (centroid_x+80,centroid_y-20),(255,0,0),1)
                    cv2.putText(image ,'speedx('+str(speed2x)+')', (centroid_x+100,centroid_y), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                elif speed2x <0:
                    cv2.line(image ,(centroid_x,centroid_y),(centroid_x-100 ,centroid_y),(255,0,0),1)
                    cv2.line(image ,(centroid_x-100,centroid_y),(centroid_x-80 ,centroid_y+20),(255,0,0),1)
                    cv2.line(image ,(centroid_x-100,centroid_y),(centroid_x-80 ,centroid_y-20),(255,0,0),1)
                    cv2.putText(image ,'speedx('+str(speed2x)+')', (centroid_x-100,centroid_y), cv2.FONT_HERSHEY_PLAIN ,1 ,(100,0,0) ,1)
                
                d=x2
                b=y2
                
            n=n+1
        else :
            pass
        
        
        
    def cont(contours , image):
        if cv2.contourArea(contours)>1000:
            cv2.drawContours(image, contours,-1,(255,0,0) , 2)
        else :
            pass
        
    edg = cv2.Canny(threshold , 10 ,180)
    dup =edg.copy()
    contours ,hierarchy = cv2.findContours(dup ,cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
    
    for i in contours:
        cont(i , black_image)
        cent(i , black_image)
        
        
    cv2.imshow('vedio', black_image)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
        
cap.release()       
cv2.destroyAllWindows()
```
